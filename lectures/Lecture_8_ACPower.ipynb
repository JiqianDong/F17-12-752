```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Vmax = 120*np.sqrt(2) # Volts
Imax = 100 # Amperes

f = 60 # frequency in Hz
w = 2*np.pi*f # frequency in radians/s
phi = 0 # * np.pi # Phase difference, in radians 


t = np.linspace(0,0.016666,num=1000)

i = Imax*np.sin(w*t)
v = Vmax*np.sin(w*t+phi)
p = i*v
```

```python
plt.plot(p)
```

```python

f1 = plt.figure(0)
plt.plot(t,v,'.r')
plt.plot(t,i,'.b')

f2 = plt.figure(1)
plt.plot(t,p,'.g')

P = Vmax*Imax/2*np.cos(phi)
Q = Vmax*Imax/2*np.sin(phi)

print("Real power is really {0: f}".format(P))
print("Reactive power is really {0: f}".format(Q))
print("Power factor is: {0: f}".format(np.cos(phi)))
print("Power factor is really {0: f}".format(P/np.sqrt(P**2+Q**2)))

```
